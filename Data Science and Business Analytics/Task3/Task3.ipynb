```python
# Importing all the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# To ignore the warnings
import warnings as wg
wg.filterwarnings("ignore")
```

```python
# Assigning path of the file 
path = "/content/drive/MyDrive/SampleSuperstore.csv"
```

```python
# Reading data iris dataset
ds = pd.read_csv(path)
```

```python
# Dropping the duplicate values
ds.drop_duplicates()
ds.head()
```

```python
# Removing the unnecessary columns such as postal code
ds = ds.drop(['Postal Code'],axis=1)
```

```python
ds.head()
```

```python
# Visualising the dataset as a ehole using the pair plot
import seaborn as sb
sb.pairplot(ds)
```

```python
# Finding the pairwise correlations between the columns and visualising using heatmaps
ds.corr()
plt.figure(figsize = (10,5))
sb.heatmap(ds.corr(), annot=True)
plt.show
```

```python
# Visualising the categories
plt.figure(figsize = (6,6))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(ds['Category'].value_counts(), labels = ds['Category'].value_counts().index, autopct = '%1.1f%%', textprops = textprops)
plt.show()
```

```python
plt.figure(figsize = (10,16))
ds.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
# Computing top categories in terms of sales from first 100 observations
top_category_sales = ds.groupby("Category").Sales.sum().nlargest(n=100)
# Computing top categories in terms of Profit from first 100 observations
top_category_profit = ds.groupby("Category").Profit.sum().nlargest(n=100)

# Plotting to see it visually
plt.style.use('seaborn')
top_category_sales.plot(kind = 'bar', figsize= (10,5), fontsize =14)
top_category_profit.plot(kind = 'bar', figsize= (10,5), fontsize =14, color='red')
plt.xlabel('Category' , fontsize = 15)
plt.ylabel('Total Sales/Profits' , fontsize=15)
plt.title("Top Category Sales vs Profit" , fontsize=15)
plt.show()
```

```python
# Computing top sub-categories in terms of sales from first 100 observations
top_subcategory_sales = ds.groupby("Sub-Category").Sales.sum().nlargest(n=100)
# Computing top sub-categories in terms of Profit from first 100 observations
top_subcategory_profit = ds.groupby("Sub-Category").Profit.sum().nlargest(n=100)

# Plotting to see it visually
plt.style.use('seaborn')
top_subcategory_sales.plot(kind = 'bar', figsize= (10,5), fontsize =14)
top_subcategory_profit.plot(kind = 'bar', figsize= (10,5), fontsize =14, color='red')
plt.xlabel('Sub-Category' , fontsize = 15)
plt.ylabel('Total Sales/Profits' , fontsize=15)
plt.title("Top Sub-Category Sales vs Profit" , fontsize=15)
plt.show()
```

```python
# A more detailed view
plt.figure(figsize=(14,12))
statewise = ds.groupby(['Sub-Category']) ['Profit'].sum().nlargest(50)
statewise.plot.barh() # h for horizontal
```

The above graph clearly shows that Copiers and Phones have the highest sales and profit and tables has negative profit

```python
# Visualising the discount
plt.figure(figsize=(8,7))
sb.lineplot(ds['Discount'], ds['Profit'], data = ds)
```

```python
# Visualising the sales vs profits in different regions
plt.figure(figsize=(6,6))
plt.title('Region')
plt.pie(ds['Region'].value_counts(), labels = ds['Region'].value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
# Computing top State in terms of sales from first 100 observations
top_state_sales = ds.groupby("State").Sales.sum().nlargest(n=100)
# Computing top State in terms of Profit from first 100 observations
top_state_profit = ds.groupby("State").Profit.sum().nlargest(n=100)

# Plotting to see it visually
plt.style.use('seaborn')
top_subcategory_sales.plot(kind = 'bar', figsize= (10,5), fontsize =14)
top_subcategory_profit.plot(kind = 'bar', figsize= (10,5), fontsize =14, color='red')
plt.xlabel('State' , fontsize = 15)
plt.ylabel('Total Sales/Profits' , fontsize=15)
plt.title("Top State Sales vs Profit" , fontsize=15)
plt.show()
```

```python
#Checking the interdependency of sales, profit and discount

plt.style.use('seaborn')
ds.plot(kind = "scatter",figsize= (10,5), x = "Sales", y="Profit", c= "Discount", s=20,fontsize=16, colormap="viridis")
plt.ylabel('Total Profits', fontsize = 16)
plt.title("Interdependency od sales,profit and discounts" , fontsize =16)
plt.show()
```

The graph clearly shows that if we give more discount on our products sales increases but profit decreases
